```python
def most_repeaed_spam():
    words = []
    words_separated = []
    dictionary = dict()
    for sentence in spam_frame['v2']:
        words_2 = sentence.split(' ')
        words.append(words_2)
        for word in words_2:
            words.append(word)
            m = dictionary.get(word, 0)
            dictionary[word] = m + 1
    sorted_by_occurence = sorted(dictionary.items(), key=lambda x:x[1], reverse=True)
    converted_dict = dict(sorted_by_occurence)
    first30pairs = {k: converted_dict[k] for k in list(converted_dict)[:30]}
    fig, ax = plt.subplots(figsize = (16,6))
    plt.bar(first30pairs.keys(),first30pairs.values(),width=0.2,color='lime')
    ax.set_xlabel("words")
    ax.set_ylabel("occurences")
    return converted_dict
```

```python
def most_repeated_words():
    words = []
    words_separated = []
    dictionary = dict()
    for sentence in data_frame['v2']:
        words_2 = sentence.split(' ')
        words.append(words_2)
        for word in words_2:
            words.append(word)
            m = dictionary.get(word, 0)
            dictionary[word] = m + 1
    for i in range(data_frame.shape[0]):
        if not pd.isna(data_frame['Unnamed: 2'][i]):
            sentence = data_frame['Unnamed: 2'][i]
            words_2 = sentence.split(' ')
            words.append(words_2)
            for word in words_2:
                words.append(word)
                m = dictionary.get(word, 0)
                dictionary[word] = m + 1
    for i in range(data_frame.shape[0]):
        if not pd.isna(data_frame['Unnamed: 3'][i]):
            sentence = data_frame['Unnamed: 3'][i]
            words_2 = sentence.split(' ')
            words.append(words_2)
            for word in words_2:
                words.append(word)
                m = dictionary.get(word, 0)
                dictionary[word] = m + 1
    for i in range(data_frame.shape[0]):
        if not pd.isna(data_frame['Unnamed: 4'][i]):
            sentence = data_frame['Unnamed: 4'][i]
            words_2 = sentence.split(' ')
            words.append(words_2)
            for word in words_2:
                words.append(word)
                m = dictionary.get(word, 0)
                dictionary[word] = m + 1
    sorted_by_occurence = sorted(dictionary.items(), key=lambda x:x[1], reverse=True)
    converted_dict = dict(sorted_by_occurence)
    first20pairs = {k: converted_dict[k] for k in list(converted_dict)[:20]}
    return first20pairs
```

```python
def length_intervals(x):
    if x < 50:
        return "less than 50"
    elif x >= 50 and x < 100:
        return "between 50 and 100"
    elif x >= 100 and x < 200:
        return "between 100 and 200"
    elif x >= 200 and x < 300:
        return "between 200 and 300"
    elif x >= 300 and x < 400:
        return "between 300 and 400"
    else:
        return "greater than 400"
    
```

```python
data_frame['length_intervals'] = data_frame['length'].map(lambda x : length_intervals(x))
spam_frame['length_intervals'] = spam_frame['length'].map(lambda x : length_intervals(x))
ham_frame['length_intervals'] = ham_frame['length'].map(lambda x : length_intervals(x))
```

```python
def plot_len_vs_count():
    g = sns.countplot(
        y= data_frame['length_intervals'],
        data = data_frame,
        order = ["less than 50","between 50 and 100","between 100 and 200","between 200 and 300","between 300 and 400","greater than 400"],
        hue = data_frame['result']
    )
    g.set_title('length vs Number of spam', fontdict = { 'fontsize': 16, 'fontweight':'bold'})
    g.set_xlabel('Count', fontsize=15, fontweight='bold')
    g.set_ylabel('length', fontsize=15, fontweight='bold')
    g.legend(loc='center right')
```

```python
def calculate_digits_num():
    counts = []
    for sentence in data_frame['v2']:
        count = 0
        for word in sentence:
            for ch in word:
                if ch.isdigit():
                    count += 1
        counts.append(count)
    data_frame['digits'] = counts
```

```python
def digits_intervals(x):
    if x <= 1:
        return "between 0 and 1"
    elif x > 1 and x < 10:
        return "between 1 and 10"
    elif x >= 10 and x < 30:
        return "between 10 and 30"
    else:
        return "greater than 30"
```

```python
data_frame['dig_ints'] = data_frame['digits'].map(lambda x : digits_intervals(x))
```

```python
def plot_digs_vs_count():
    g = sns.countplot(
        y= data_frame['dig_ints'],
        data = data_frame,
        order = ["between 0 and 1","between 1 and 10","between 10 and 30","greater than 30"],
        hue = data_frame['result']
    )
    g.set_title('digits count vs Number of spam', fontdict = { 'fontsize': 16, 'fontweight':'bold'})
    g.set_xlabel('digit count', fontsize=15, fontweight='bold')
    g.set_ylabel('spam & ham count', fontsize=15, fontweight='bold')
    g.legend(loc='center right')
```

```python
def update_length():
    lengths = []
    for i in range(data_frame.shape[0]):
        count = 0
        if not pd.isna(data_frame['Unnamed: 2'][i]):
            count += len(data_frame['Unnamed: 2'][i])
        if not pd.isna(data_frame['Unnamed: 3'][i]):
            count += len(data_frame['Unnamed: 3'][i])
        if not pd.isna(data_frame['Unnamed: 4'][i]):
            count += len(data_frame['Unnamed: 4'][i])
        lengths.append(data_frame['length'][i] + count)
    data_frame['length'] = lengths
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import  train_test_split
```

```python
new_data_frame = data_frame.drop(['length_intervals','dig_ints'], axis = 1)
```

```python
count = CountVectorizer()
X = new_data_frame.drop('result', axis = 1)
y = new_data_frame['result']
msgs = count.fit_transform(X['v2'])

X_train , X_test, y_train, y_test = train_test_split(msgs,y,test_size=0.2,random_state=123)
```

```python
msgs
```

```python
X_test
```

```python
from sklearn.naive_bayes import BernoulliNB

classifier = BernoulliNB()  
classifier.fit(X_train,y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
from sklearn.metrics import classification_report
print('Classification report:\n', classification_report(y_test, y_pred))
```
